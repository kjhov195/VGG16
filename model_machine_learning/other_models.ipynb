```python
def process_image(image_path):
    global img
    """Process an image path into a PyTorch tensor"""

    image = Image.open(image_path)
    # Resize
    img = image.resize((256, 256))

    # Center crop
    width = 256
    height = 256
    new_width = 224
    new_height = 224

    left = (width - new_width) / 2
    top = (height - new_height) / 2
    right = (width + new_width) / 2
    bottom = (height + new_height) / 2
    img = img.crop((left, top, right, bottom))

    # Convert to numpy, transpose color dimension and normalize
    img = np.array(img).transpose((2, 0, 1)) / 256

    # Standardization
    means = np.array([0.485, 0.456, 0.406]).reshape((3, 1, 1))
    stds = np.array([0.229, 0.224, 0.225]).reshape((3, 1, 1))

    img = img - means
    img = img / stds

    #img_tensor = torch.Tensor(img)

    #return img_tensor
    return img
```

```python
list_train = []
list_valid = []
list_test = []
```

```python
list_y_label = []
list_y_testlabel = []
```

```python
whattodo = 'train'
obj_list = ['car','bus','truck']

for obj in obj_list:
    path2 = './dataset/%s/%s/'%(whattodo, obj)
    for file_name in os.listdir(path2):
        try:
            list_train.append(process_image(path2+file_name).flatten())
            list_y_label.append(obj)
            print(file_name)
        except ValueError:
            pass
```

```python
print(len(list_train), len(list_y_label))
```

```python
whattodo = 'test'
obj_list = ['car','bus','truck']

for obj in obj_list:
    path2 = './dataset/%s/%s/'%(whattodo, obj)
    for file_name in os.listdir(path2):
        try:
            list_test.append(process_image(path2+file_name).flatten())
            list_y_testlabel.append(obj)
            print(file_name)
        except ValueError:
            pass
```

# KNN prediction

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
img_data = list_train
img_label_orig = list_y_label
sample_test_data = list_test
```

```python
nbrs = KNeighborsClassifier(n_neighbors=3, algorithm='brute').fit(img_data, img_label_orig)
YPred_KNN = nbrs.predict(sample_test_data)
```

```python
A = np.array(list_y_testlabel)
```

```python
B = np.array(YPred_KNN)
```

```python
for items in obj_list:
    print("accuracy of %s class: %f"%(items, sum(B[A==items]==items)/sum(A==items)))
```
